```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## On Given Dataset

```python
df=pd.read_csv('Data1.csv')
df
```

```python
plt.scatter(df.Week,df.Sales)
plt.xlabel("Week")
plt.ylabel("Sales in Thousands")
for i in range(len(df)):
    plt.annotate(f"({df.Week[i]}, {df.Sales[i]})", xy=(df.Week[i], df.Sales[i]), xycoords='data', textcoords='data')
```

```python
x=df[['Week']]
y=df[['Sales']]
reg=LinearRegression()

reg.fit(x,y)
a1=reg.coef_
a0=reg.intercept_
y1=a0+a1*x
print("a0=",a0,", a1=",a1)
```

***Regression Expression: y = a0+a1x = 0.66+0.54x***

```python
plt.scatter(x,y)
plt.plot(x,y1,'r')
plt.xlabel("Week")
plt.ylabel("Sales in Thousands")
for i in range(len(df)):
    plt.annotate(f"({df.Week[i]}, {df.Sales[i]})", xy=(df.Week[i], df.Sales[i]), xycoords='data', textcoords='data')
```

```python
y7=reg.predict([[7]])
y12=reg.predict([[12]])
print("Predicted 7th week sale=",y7)
print("Predicted 12th week sale=",y12)
```

## On Random Floating Number Sets

```python
xf=np.random.uniform(low=0.0,high=1.0,size=10)
xfr=xf.reshape(-1, 1)
yf=np.random.uniform(low=0.0,high=1.0,size=10)
yfr=yf.reshape(-1, 1)
```

```python
plt.scatter(xf,yf)
plt.xlabel("Week")
plt.ylabel("Sales in Thousands")
for i in range(len(xf)):
    plt.annotate(f"({xf[i].round(2)}, {yf[i].round(2)})", xy=(xf[i], yf[i]), xycoords='data', textcoords='data')
```

```python
reg.fit(xfr,yfr)
a1f=reg.coef_
a0f=reg.intercept_
y3=a0f+a1f*xf
print("a0=",a0f,", a1=",a1f)
```

***Regression Expression: y = a0+a1x***

```python
plt.scatter(xf,yf)
sns.lineplot(x=xf,y=y3.reshape(-1),color='red')
plt.xlabel("Week")
plt.ylabel("Sales in Thousands")
for i in range(len(xf)):
    plt.annotate(f"({xf[i].round(2)}, {yf[i].round(2)})", xy=(xf[i], yf[i]), xycoords='data', textcoords='data')
```

```python
y7f=reg.predict([[7]])
y12f=reg.predict([[12]])
print("Predicted 7th week sale=",y7f)
print("Predicted 12th week sale=",y12f)
```
