```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## On Given Dataset

```python
df=pd.read_excel('multidata.xlsx')
df
```

```python
x1=np.array(df.x1)
x2=np.array(df.x2)
y=np.array(df.y)
```

```python
plt.scatter(x1,y,color='red',label='X1 v Y')
plt.scatter(x2,y,color='blue',marker='x',label='X2 v Y')
plt.legend()
for i in range(len(x1)):
    plt.annotate(f"({x1[i]}, {y[i]})", xy=(x1[i], y[i]), xycoords='data', textcoords='data')
    plt.annotate(f"({x2[i]}, {y[i]})", xy=(x2[i], y[i]), xycoords='data', textcoords='data')
```

```python
n=len(df)
x12=x1*x1
x22=x2*x2
x1x2=x1*x2
x1y=x1*y
x2y=x2*y
x12s=x12.sum()-x1.sum()*x1.sum()/n
x22s=x22.sum()-x2.sum()*x2.sum()/n
x1x2s=x1x2.sum()-x1.sum()*x2.sum()/n
x1ys=x1y.sum()-x1.sum()*y.sum()/n
x2ys=x2y.sum()-x2.sum()*y.sum()/n

b1=(x22s*x1ys-x1x2s*x2ys)/(x12s*x22s-x1x2s*x1x2s)
b2=(x12s*x2ys-x1x2s*x1ys)/(x12s*x22s-x1x2s*x1x2s)
a=y.mean()-b1*x1.mean()-b2*x2.mean()

yc=a+b1*x1+b2*x2

print("a=",a,", b1=",b1,", b2=",b2)
```

***Regression Expression: Y = a+b1x1+b2x2 = 2.799+2.28x1-1.67x2***

```python
plt.scatter(x1,y,color='red',label='X1 v Y')
plt.scatter(x2,y,color='blue',marker='x',label='X2 v Y')
plt.plot(yc,yc,'g',label='Model')
plt.legend()
for i in range(len(x1)):
    plt.annotate(f"({x1[i]}, {y[i]})", xy=(x1[i], y[i]), xycoords='data', textcoords='data')
    plt.annotate(f"({x2[i]}, {y[i]})", xy=(x2[i], y[i]), xycoords='data', textcoords='data')
```

```python
y321=a+b1*3+b2*2
print("Prediction for x1=3 and x2=2 =",y321)
```

## On Random Integer Sets

```python
x1i=np.random.randint(1,20,size=10)
x2i=np.random.randint(1,20,size=10)
yi=np.random.randint(1,20,size=10)
```

```python
plt.scatter(x1i,yi,color='red',label='X1 v Y')
plt.scatter(x2i,yi,color='blue',marker='x',label='X2 v Y')
plt.legend()
for i in range(len(x1i)):
    plt.annotate(f"({x1i[i]}, {yi[i]})", xy=(x1i[i], yi[i]), xycoords='data', textcoords='data')
    plt.annotate(f"({x2i[i]}, {yi[i]})", xy=(x2i[i], yi[i]), xycoords='data', textcoords='data')
```

```python
ni=len(x1i)
x12i=x1i*x1i
x22i=x2i*x2i
x1x2i=x1i*x2i
x1yi=x1i*yi
x2yi=x2i*yi
x12si=x12i.sum()-x1i.sum()*x1i.sum()/ni
x22si=x22i.sum()-x2i.sum()*x2i.sum()/ni
x1x2si=x1x2i.sum()-x1i.sum()*x2i.sum()/ni
x1ysi=x1yi.sum()-x1i.sum()*yi.sum()/ni
x2ysi=x2yi.sum()-x2i.sum()*yi.sum()/ni

b1i=(x22si*x1ysi-x1x2si*x2ysi)/(x12si*x22si-x1x2si*x1x2si)
b2i=(x12si*x2ysi-x1x2si*x1ysi)/(x12si*x22si-x1x2si*x1x2si)
ai=yi.mean()-b1i*x1i.mean()-b2i*x2i.mean()

yci=ai+b1i*x1i+b2i*x2i

print("a=",ai,", b1=",b1i,", b2=",b2i)
```

***Regression Expression: Y = a+b1x1+b2x2***

```python
plt.scatter(x1i,yi,color='red',label='X1 v Y')
plt.scatter(x2i,yi,color='blue',marker='x',label='X2 v Y')
plt.plot(yci,yci,'g',label='Model')
plt.legend()
for i in range(len(x1i)):
    plt.annotate(f"({x1i[i]}, {yi[i]})", xy=(x1i[i], yi[i]), xycoords='data', textcoords='data')
    plt.annotate(f"({x2i[i]}, {yi[i]})", xy=(x2i[i], yi[i]), xycoords='data', textcoords='data')
```

```python
y32i=ai+b1i*3+b2i*2
print("Prediction for x1=3 and x2=2 =",y32i)
```

## On Random Floating Number Sets

```python
x1f=np.random.uniform(1,20,size=10).round(2)
x2f=np.random.uniform(1,20,size=10).round(2)
yf=np.random.uniform(1,20,size=10).round(2)
```

```python
plt.scatter(x1f,yf,color='red',label='X1 v Y')
plt.scatter(x2f,yf,color='blue',marker='x',label='X2 v Y')
plt.legend()
for i in range(len(x1f)):
    plt.annotate(f"({x1f[i].round(2)}, {yf[i].round(2)})", xy=(x1f[i], yf[i]), xycoords='data', textcoords='data')
    plt.annotate(f"({x2f[i].round(2)}, {yf[i].round(2)})", xy=(x2f[i], yf[i]), xycoords='data', textcoords='data')
```

```python
nf=len(x1f)
x12f=x1f*x1f
x22f=x2f*x2f
x1x2f=x1f*x2f
x1yf=x1f*yf
x2yf=x2f*yf
x12sf=x12f.sum()-x1f.sum()*x1f.sum()/nf
x22sf=x22f.sum()-x2f.sum()*x2f.sum()/nf
x1x2sf=x1x2f.sum()-x1f.sum()*x2f.sum()/nf
x1ysf=x1yf.sum()-x1f.sum()*yf.sum()/nf
x2ysf=x2yf.sum()-x2f.sum()*yf.sum()/nf

b1f=(x22sf*x1ysf-x1x2sf*x2ysf)/(x12sf*x22sf-x1x2sf*x1x2sf)
b2f=(x12sf*x2ysf-x1x2sf*x1ysf)/(x12sf*x22sf-x1x2sf*x1x2sf)
af=yf.mean()-b1f*x1f.mean()-b2f*x2f.mean()

ycf=af+b1f*x1f+b2f*x2f

print("a=",af,", b1=",b1f,", b2=",b2f)
```

***Regression Expression: Y = a+b1x1+b2x2***

```python
plt.scatter(x1f,yf,color='red',label='X1 v Y')
plt.scatter(x2f,yf,color='blue',marker='x',label='X2 v Y')
plt.plot(ycf,ycf,'g',label='Model')
plt.legend()
for i in range(len(x1f)):
    plt.annotate(f"({x1f[i].round(2)}, {yf[i].round(2)})", xy=(x1f[i], yf[i]), xycoords='data', textcoords='data')
    plt.annotate(f"({x2f[i].round(2)}, {yf[i].round(2)})", xy=(x2f[i], yf[i]), xycoords='data', textcoords='data')
```

```python
y32f=af+b1f*3+b2f*2
print("Prediction for x1=3 and x2=2 =",y32f)
```
