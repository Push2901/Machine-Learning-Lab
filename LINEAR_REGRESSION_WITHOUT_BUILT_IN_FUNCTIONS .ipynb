## Without using built-in functions

```python
sample_data: pd.DataFrame = pd.read_csv("Data1.csv")
sample_data
```

**2.** Plot a scatter plot of the data.

```python
sample_data.plot.scatter("Week", "Sales")
```

**3.** Find the best fit line using linear regression.

```python
def find_regression_line (df: pd.DataFrame) -> tuple[float, float]:
    if (len(df.columns) != 2):
        raise Exception("Regression data must have 2 columns.")

    copy_df = df.copy(deep=True).rename(columns={
        df.columns[0]: "x",
        df.columns[1]: "y"
    })

    copy_df["x2"] = copy_df.x ** 2
    copy_df["xy"] = copy_df.x * copy_df.y
    
    x = copy_df.x.mean()
    y = copy_df.y.mean()
    a1 = (copy_df.xy.mean() - x*y) / (copy_df.x2.mean() - x**2)
    a0 = y - a1*x
    
    return (a1, a0)
```

```python
(a1, a0) = find_regression_line(sample_data)
print(f"Required line: y = ({a0}) + ({a1})x")
```

**4.** Fit line in the scatter plot.

```python
sample_data.plot.scatter("Week", "Sales")

x: np.ndarray = np.linspace(sample_data.Week.min(), sample_data.Week.max())
y: np.ndarray = a0 + a1*x
plt.plot(x, y)

plt.title("Fitting regression curve")
plt.show()
```

**5.** Predict sales for weeks 7 and 12.

```python
sample_data.loc[len(sample_data.index)] = [7, a0 + a1*7]
sample_data.loc[len(sample_data.index)] = [12, a0 + a1*12]
sample_data
```

**6.** Plot the points in a graph showing the coordinates.

```python
sample_data.plot.scatter(x="Week", y="Sales")

for x, y in zip(sample_data.Week, sample_data.Sales):
    plt.annotate('(%.2f, %.2f)' % (x, y), xy=(x, y), textcoords="offset points", xytext=(0, 10), ha="center")
    
x: np.ndarray = np.linspace(sample_data.Week.min(), sample_data.Week.max())
y: np.ndarray = a0 + a1*x
plt.plot(x, y)
```
